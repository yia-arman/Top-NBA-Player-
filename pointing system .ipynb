### Creating a pointing system 
        - 2 point for the top 5 
        - 1 point for the top 10 
        - 0 point for the rest  

```python
import pandas as pd
```

```python
df = pd.read_csv('player_info.csv')
df
```

```python
df = df.drop([3,8,14])
df
```

```python
df = df.reset_index(drop=True)
```

```python
df['Point'] = 0
df_copy = df.copy()
```

```python
df["Point"] += df.G.apply(lambda x: 2 if x>55 else 1 if x>=50 else 0)
df["Point"] += df.PTS.apply(lambda x: 2 if x>30 else 1 if x>=25 else 0)
df["Point"] += df.AST.apply(lambda x: 2 if x>6.8 else 1 if x>=5 else 0)
df["Point"] += df.TRB.apply(lambda x: 2 if x>10 else 1 if x>=7 else 0)
df["Point"] += df.STL.apply(lambda x: 2 if x>1.3 else 1 if x>=1 else 0)
df["Point"] += df.BLK.apply(lambda x: 2 if x>1 else 1 if x>=0.6 else 0)
df["Point"] += df.TOV.apply(lambda x: -2 if x>3.2 else -1 if x>=2.7 else 0)
df["Point"] += df.PF.apply(lambda x: -2 if x>3 else -1 if x>=2.2 else 0)
```

```python
new_df = df.sort_values('Point', ascending=False)
new_df = new_df.reset_index(drop=True)
```

```python
new_df = new_df[['Player', 'Team','Salary', 'Point']]
```
